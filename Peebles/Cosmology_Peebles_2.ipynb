```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
from scipy import integrate as integrate
```

```python
plt.style.use('default')
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
def E(z, omega, omega_R, omega_lambda):
    return ((omega*((1+z)**3)) +
            (omega_R*((1+z)**2))+ omega_lambda)**(1/2)
```

```python
def integral(f,z,omega,omega_R, omega_lambda):
    
    I = integrate.quad(f, 0, z, args=(omega,omega_R, omega_lambda,))[0]
    return I
```

# Probability for intersetcing object along a line of sight (Fig 13.4 Peebles)

```python
def dPdz(z,omega,omega_R,omega_lambda):
    return (((1+z)**2)/(E(z, omega, omega_R, omega_lambda)))
    
```

```python
def dFglzs(z,zs,omega):
    prefactor = 16*(np.pi)**3/(E(z,omega,1-omega,0))
    yol = Mattig_y(omega,z)
    yos = Mattig_y(omega,zs)
    yls = yos*((1+((yol)**2)*(1-omega))**(1/2))-yol*((1+((yos)**2)*(1-omega))**(1/2))
    result = prefactor* ((yol*yls/yos)**2)
    return result
```

```python
def Fglzs(dFglzs,zs,omega):
    
    I = integrate.quad(dFglzs, 0, zs, args=(zs,omega,))[0]
    return I
```

```python
def dDs(z,omega,omega_R,omega_lambda):
    return (5*omega*(1+z))/(2*((E(z, omega, omega_R, omega_lambda))**3))
```

```python
def Ds(dDs,omega,omega_R, omega_lambda):
    
    I = integrate.quad(dDs, 0, np.inf, args=(omega,omega_R, omega_lambda,))[0]
    return I
```
